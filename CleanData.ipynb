## Create a new Datarame only for Countries

```python
countriesdf = df.filter(['Country Name','Country Code'], axis=1)
countriesdf=countriesdf.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
countriesdf['index_country'] = range(1, len(countriesdf) + 1)
countriesdf
```

```python
countriesdf=countriesdf.loc[:, ["index_country","Country Name","Country Code"]]
countriesdf
```

## Create a new Datarame only for Indicators

```python
indicatorsdf = df.filter(['Indicator Name','Indicator Code'], axis=1)
indicatorsdf=indicatorsdf.drop_duplicates(subset=None, keep='first', inplace=False)
indicatorsdf
```

```python
indicatorsdf['index_indicat'] = range(1, len(indicatorsdf) + 1)
indicatorsdf
```

```python
indicatorsdf=indicatorsdf.loc[:, ["index_indicat","Indicator Name","Indicator Code"]]
indicatorsdf
```

## Merge 2 Tables into a new one with indexes

```python
a = pd.merge(countriesdf,df,how='inner',on='Country Name')
a = a.drop('Country Code_y', 1)
```

```python
a = pd.merge(indicatorsdf,a,how='inner',on='Indicator Name')
```
